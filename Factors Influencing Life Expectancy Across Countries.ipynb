### 3. Data Preprocessing

This section covers data cleaning steps such as handling missing values, removing duplicates, converting data types, and creating new useful columns.

```python
# This line checks for missing (null) values in each column of the dataset by counting how many null entries are present per variable.
df.isnull().sum()
```

The output lists each column along with the total number of missing values it contains. This helps identify:

* Columns with no missing values

* Columns that require data cleaning or imputation

* The overall extent of missing data in the dataset

Understanding missing values is essential for deciding appropriate preprocessing techniques such as removal, imputation, or transformation.

```python
# Numerical columns: fill missing values with median
# Categorical columns: fill missing with mode

num_cols = df.select_dtypes(include=['float64', 'int64']).columns
cat_cols = df.select_dtypes(include=['object']).columns

df[num_cols] = df[num_cols].fillna(df[num_cols].median())
df[cat_cols] = df[cat_cols].fillna(df[cat_cols].mode().iloc[0])
```

* Numerical and categorical features are handled separately using appropriate imputation strategies.

* Median imputation minimizes the impact of outliers on numerical data.

* Mode imputation preserves dominant categories in categorical data.

* The dataset is now free of missing values and ready for further analysis and modeling.

This preprocessing step ensures data quality and integrity by systematically addressing missing values across both numerical and categorical variables. By applying suitable imputation techniques based on data type, the dataset becomes complete and reliable, reducing bias and improving the accuracy of subsequent exploratory analysis and predictive modeling.

```python
df.isnull().sum()
```

The output shows zero missing values across all columns, confirming that the dataset is now complete and free from null entries. This verification step ensures the data is clean and ready for reliable exploratory analysis and modeling.

#### Removing Duplicates

```python
# Remove duplicated rows
df = df.drop_duplicates()

df.shape
```

All duplicate records, if present, are removed, improving data accuracy and preventing duplicate observations from biasing the analysis.

The output displays the new number of rows and columns, confirming whether duplicates were found and removed. A reduction in row count indicates the presence of duplicate records in the original dataset.

```python
# Convert 'Year' to integer if needed
if df['Year'].dtype != 'int64':
    df['Year'] = df['Year'].astype(int)

# Strip whitespace from categorical fields
for col in cat_cols:
    df[col] = df[col].str.strip()

df.dtypes
```

* The Year column is now stored as an integer, enabling accurate chronological sorting, grouping, and time-series analysis.

* All categorical columns are cleaned of unnecessary whitespace, ensuring consistent category labels.

* The displayed data types confirm that each column is now in an appropriate and standardized format.

* These steps help prevent hidden data quality issues that could affect grouping, filtering, or modeling results.

#### Creating Derived Columns

```python
# Example derived column GDP_per_Capita:
if 'GDP' in df.columns and 'Population' in df.columns:
    df['GDP_per_Capita'] = df['GDP'] / df['Population']
```

* A new feature, GDP per Capita, is successfully created when both required columns are available.

* This variable allows better comparison of economic well-being across countries with different population sizes.

* GDP per capita is expected to show a stronger relationship with life expectancy than total GDP, making it valuable for further analysis.

By creating the GDP_per_Capita feature, the dataset is enriched with a more informative economic indicator. This transformation enhances analytical depth and supports more accurate insights into the relationship between economic conditions and life expectancy.

#### Filtering/Aggregating Data

```python
# Filter out rows with unrealistic Life Expectancy values
df.columns = df.columns.str.strip()

df = df[(df['Life expectancy'] > 0) & (df['Life expectancy'] <= 100)]
df.shape
```

* Column names are standardized, ensuring consistent and error-free referencing.

* Records with unrealistic life expectancy values (≤ 0 or > 100 years) are removed.

* The dataset size is reduced only to include valid and meaningful observations.

* This step improves the credibility and reliability of subsequent analysis.

By cleaning column names and filtering unrealistic life expectancy values, this step enhances data quality and ensures that the analysis is based on valid, real-world observations. This validation process is essential for producing accurate insights and trustworthy conclusions in life expectancy analysis.
